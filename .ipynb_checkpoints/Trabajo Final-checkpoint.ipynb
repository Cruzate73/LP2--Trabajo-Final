```python
# LIBRERÍAS:
import pandas as pd # PAQUETE PANDAS PARA CREAR UN DATAFRAME.
from selenium import webdriver # PARA LEER LA DESCARGA DEL DRIVER.
from selenium.webdriver.support.ui import WebDriverWait  # NOS PERMITE ESPERAR ESPERAR HASTA QUE TODOS LOS ELEMENTOS DE 
# LA PÁGINA WEB CARGUE Y RECIEN AHÍ DARA LUZ VERDE PARA QUE SE EJECUTE LAS ACCIONES PROGRAMADAS EN LAS LÍNEAS DE CÓDIGO.
from selenium.webdriver.support import expected_conditions as EC  # PARA GENERAR UNA CONDICIÓN DE EXTRACCIÓN.
from selenium.webdriver.common.by import By  # PARA PODER LEER ELEMENTOS DEL HTML.
import numpy as np # PAQUETE numpy PARA CREAR DATAFRAME.

# Opciones de navegación:
# con estas líeas de código solo buscamos que la ventana de chrome al momento
# de abrirse muestre la pantalla completa con todas las extenciones de chrome y no sea una pagina limitada.
options= webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--start-extensions')

# LECTURA DEL DRIVER:
path = 'C:/Users/user/Desktop/ESCRITORIO/LP2/Trabajo Final/chromedriver'# ESTE ES LA DIRECCIÓN DONDE SE GUARDO LA DESCARGA DEL DRIVER QUE UTUILIZAREMOS PARA LEER LA PAGINA.
driver = webdriver.Chrome(path, chrome_options=options) # ESTO SIRVE PARA ACTIVAR EL DRIVER Y PODAMOS LEER LA PAGÍNA.


# PARA BUSCAR EL GEN REQUERIDO:
lista_de_genes1=['COL6A2','FBXO25','DIP2A','PLCL2','KCNH8']# LISTA PRUEBA
#lista_de_genes2=['ZBTB40','EPHB2','LUZP1','RPL11','SPOCD1']# LISTA PRUEBA
#lista_de_genes3=['GIGYF2','FOXJ3','ERMAP','TMEM125','ELOVL1']# LISTA PRUEBA

for gen in lista_de_genes1:# CREACIÓN DE UN BUCLE PARA QUE SE PUEDA BUSCAR UN DE UNA LISTA.
#    gen='ELOVL1' # SOLO GENES QUE LA PÁGINA TENGA EN SU BASE DE DATOS EN CASO CONTRARIO ARROJARA ERROR.
    website = 'https://www.uniprot.org/uniprot/?query={}'# ESTE ES EL URL DONDE PONDREMOS EL GEN REQUERIDO Y LO PUEDA BUSCAR.
    url=website.format(gen) # PARA BUSCAR EL REQUERIDO.
    driver.get(url)# CON ESTE CÓDIGO PODEMOS MOSTRAR LA PAGINA WEB CON EL GEN BUSCADO.

# PARA SELECCIONAR LAS OPCIONES QUE NOS PERMITIRA OBTENER LA INFORMACIÓN PEDIDA.
# Cuando buscamos el gen la página web nos arroja información basica pero tiene una 
# opción al costado que nos permite escojer que información queremos del gen por lo que 
# para no complicarnos al momento de recopilar la información hemos usado una serie de códgos que nos
# ayuda a poder seleccionar que información queremos del gen y asi poder facilitarnos el trabajo 
# debido a que si entramos a la base de datos completa nos emorariamos en obtener la información por eso opamos 
# por esta opción que nos brinda la página web.

# CÓDIGO
# Le estamos diciendo al driver que tiene 5 segundos para encontrar la página y que ejecute lo pedidos en las líneas que siguen.
# Creamos una condición de busqueda
# Una vez que inspeccionamos lo que queremos con "By" podemos decirle a la condición que lea lo inpeccionado.
# .click() Esto es la opción para dar click que es lo que estamos buscando pero hay más opciones dependiendo lo que queramos hacer.
# .send_keys(' ') En este caso usamos esta opción debido a que queremos que la oopción que hemos selecionado en la línea de código sea quitada.

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'th.addRemoveColumn.tooltipped')))\
    .click()

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'input#reviewed-column-input')))\
    .send_keys(' ')

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'input#existence-column-input')))\
    .click()

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'input#comment_8_FUNCTION_9_-column-input')))\
    .click()

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'button#top-customize-columns-save')))\
    .click()

# PARA EXTRAER LA INFORMACIÓN REQUERIDA:
# Para esta parte usamos el "xpath" que es un elemento de html que gracias a "By" lo podemos extraer.
L=[] # Lista vacía
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[2]')))\


texto_columna1 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[2]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna1 = texto_columna1.text # Lo ponemos como texto
L.append(texto_columna1)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[3]')))\

texto_columna2 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[3]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna2 = texto_columna2.text # Lo ponemos como texto
L.append(texto_columna2)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[4]')))\

texto_columna3 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[4]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna3 = texto_columna3.text # Lo ponemos como texto
L.append(texto_columna3)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[5]')))\

texto_columna4 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[5]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna4 = texto_columna4.text # Lo ponemos como texto
L.append(texto_columna4)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[6]')))\

texto_columna5 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[6]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna5 = texto_columna5.text # Lo ponemos como texto
L.append(texto_columna5)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[7]')))\

texto_columna6 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[7]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna6 = texto_columna6.text # Lo ponemos como texto
L.append(texto_columna6)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '//html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[8]')))\

texto_columna7 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[8]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna7 = texto_columna7.text # Lo ponemos como texto
L.append(texto_columna7)


WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                     '/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[9]')))\

texto_columna8 = driver.find_element_by_xpath('/html/body/main/div/div[2]/div[2]/form/table/tbody/tr[1]/td[9]') # CON ESTA LINEA DE CÓDIGO LO QUE HACEMOS ES CREAR UNA LISTA DE TODA LA INFORMACIÓN QUE NOS PROPORCIONA LA PÁGINA CON REPECTO AL GEN BUSCADO.
texto_columna8 = texto_columna8.text # Lo ponemos como texto
L.append(texto_columna8)

# Convirtiendo la lista L en una lista de listas para poder usarlo en el dataframe
L1=[L]

# creando la dataframe:
df = pd.DataFrame(L1, columns=['Entry','Entry Name','Protein','Gene','Organism','Length','Status','Biological'])
print(df)

'''
# Creando un excel con los datos del data frame:
df.to_csv('Reporte.csv', index=False)
'''

driver.quit() # Para cerrar la página web.
```
